```python
from sklearn.model_selection import train_test_split
#Split train and test
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, shuffle=True)

#Count features for modelization
X_num_columns= len(X.columns)
```

# Creating **NN**


*  **Activition function**: The activation function of a node defines the output of that node given an input or set of inputs.
*  We created a simple NN with $3$ dense layers of the model with row_num/3, row_num/6, 1 neuron respectly,  $1$ dropping layer with $p = 0.2$ to prevent overfitting
Google provides the use of free GPU for your Colab notebooks.
Powers of 2 are your friends when training NNs on GPUs (it's based on how the GPU is designed)



```python
import numpy as np
from tensorflow.keras import layers
from keras.models import Sequential
from keras.layers import Dense,Dropout,Activation,SimpleRNN 
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt


#todo: try different number of layers 
#todo: try different try different neurons in each layer 


#todo: try RNN
#todo: try Training multilayer perceptron with Backpropagation
#todoï¼šBackprop-free reinforcement learning?


general = True
RNN = False

#Sequential Model in keras: Single input and single output. 
if general == True:
  model = Sequential([
      Dense(row_num/3, activation='relu',input_dim = X_num_columns),
      Dense(row_num/6, activation='relu'),
      Dense(1,activation='linear'),
      # adding a dropout layer
      Dropout(0.2)
      ])
  model.compile(optimizer='adam', loss='mean_squared_error')
  model.summary()
  
'''
1. Recurrent Neural Network (RNN) are types of Neural Networks 
designed to use sequential data such as time-series.
2. In RNN neural network models, we use the sigmoid or tanh function for hidden layers.
#The tanh function has better performance.
'''

if RNN == True: 
  model = Sequential([
      SimpleRNN(units=1, activation='tanh'),
      Dense(units=1, activation='tanh'),
      Dense(1,activation='linear'),
      ])
  model.compile(optimizer='adam', loss='mean_squared_error')
  model.summary()
  

```

```python
#Fit model to training data
model.fit(X_train, y_train, epochs=64, batch_size=25)

print("Training completed")
```

```python
# sum of squared errors over the sample size of 276

print(model.metrics_names)

loss = model.evaluate(X_test, y_test)

print(np.sqrt(loss))
```
